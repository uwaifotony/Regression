```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn
seaborn.set()
```

```python
data = pd.read_csv('1.02. Multiple linear regression.csv')
```

```python
data
```

```python
data.describe()
```

```python
y = data['GPA']
x1 = data[['SAT', 'Rand 1,2,3']]
```

```python
x = sm.add_constant(x1)
result = sm.OLS(y,x).fit()
```

```python
result.summary()
```
